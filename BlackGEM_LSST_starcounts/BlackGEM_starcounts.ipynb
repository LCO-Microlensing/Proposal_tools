```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
fname = "field_2229.dat"
data = pd.read_csv(fname, header=0, sep="\s+", skipfooter=1)
```

```python
data
```

```python
bins = np.arange(12, 22, 0.5)
```

```python
fig = plt.figure()
plt.hist(data["z"], bins=bins, alpha=0.6, histtype="stepfilled", color="plum", edgecolor="mediumvioletred", zorder=0, label="z");
plt.hist(data["i"], bins=bins, alpha=0.6, histtype="stepfilled", color="lightcoral", edgecolor="firebrick", zorder=0, label="i");
plt.hist(data["r"], bins=bins, alpha=0.6, histtype="stepfilled", color="tan", edgecolor="peru", zorder=0, label="r");
plt.hist(data["g"], bins=bins, alpha=0.6, histtype="stepfilled", color="limegreen", edgecolor="darkgreen", zorder=0, label="g");
plt.hist(data["u"], bins=bins, alpha=0.6, histtype="stepfilled", color="cornflowerblue", edgecolor="mediumblue", zorder=0, label="u");
plt.legend(loc="best")
plt.xlabel("Magnitude")
plt.ylabel("N stars")
plt.title("Star counts per filter using TRILEGAL simulations\n Galactic plane field, RA=245.0342, dec=-48.2973")
```
